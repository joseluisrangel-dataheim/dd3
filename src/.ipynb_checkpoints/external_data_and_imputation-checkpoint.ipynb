# Data

```python
df1 = pd.read_csv('../data/processed/metros_cubicos_data.csv')
df2 = pd.read_csv('../data/processed/scrapper_viva_data.csv')
```

```python
df1['source'] = 'metros_cubicos'
df2['source'] = 'viva_anuncios'
```

# JOIN de los datos

```python
df2 = df2.rename(columns={'price':'price_final',
                         'name':'main_name',
                         'terreno':'m2_totales'
                        
                        })
```

```python
df2['main_name'] = df2['main_name'].str.lower()
```

```python
df2['property_type'] = np.nan
df2.loc[df2.main_name.str.contains('depa|ph|depto|pent|loft'), 'property_type'] = 'depto'
df2.loc[df2.property_type.isna(), 'property_type'] = 'casa'
```

# Hacer join espacial

```python
import geopandas as gdp
import pandas as pd
```

```python
len(df)
```

```python
#Tirar los que no tienen coordenadas
df_geo = df.dropna(subset=['lon'])
```

```python
len(df_geo)
```

```python
df_gdp = gdp.GeoDataFrame(
    df_geo, 
    geometry=gdp.points_from_xy(df_geo.lon, df_geo.lat)).set_crs(epsg=4326)
```

```python
agebs_geo = gdp.read_file("../data/raw/agebs/00a.shp")  
agebs_geo = agebs_geo.to_crs(epsg=4326)
```

```python
df_geo_final = gdp.sjoin(df_gdp, agebs_geo, how="inner", op='intersects')
```

```python
df_geo_final.CVE_AGEB.unique()
```

```python
df_geo_final.CVE_MUN.value_counts()
```

```python
df_geo_final.to_csv('../data/processed/geo_data.csv', index=False)
```

```python
from dateutil import relativedelta
from shapely.geometry import mapping

import geopandas as gpd
import gmplot
```

```python
len(df_geo_final)
```

```python
df_geo_final.head(1)
```

```python
precio_medio_ageb = df_geo_final.groupby('CVEGEO').agg({'precio_m2':'median',
                                                        'precio_m2_con':'median',
                                                        'id':'count'
                                                         }).reset_index()
```

```python
precio_medio_ageb = precio_medio_ageb[precio_medio_ageb.id>5]
```

```python
agebs_geo_map = agebs_geo.copy()
agebs_geo_map = agebs_geo_map[['CVEGEO', 'geometry']]
agebs_geo_map = agebs_geo_map.drop_duplicates(subset=['CVEGEO'])
```

```python
precio_medio_ageb.head(1)
```

```python
agebs_geo_map = agebs_geo_map.merge(precio_medio_ageb, 
                                    how='left', on='CVEGEO')
```

```python
agebs_geo_map = agebs_geo_map.dropna(subset=['precio_m2'])
```

```python
agebs_geo_map = agebs_geo_map.reset_index(drop=True)
```

```python
g = [i for i in agebs_geo_map.geometry]
```

```python
agebs_geo_map['ponde'] = (
    (agebs_geo_map.precio_m2-agebs_geo_map.precio_m2.min()) /
    (agebs_geo_map.precio_m2.max()-agebs_geo_map.precio_m2.min())
)
```

```python
len(g)
```

```python
gmap = gmplot.GoogleMapPlotter(21.472548998857, -100.71556222583472, 5, apikey="")
for i in range(0, len(g)):
    all_coords = mapping(g[i])["coordinates"]
    ponde = agebs_geo_map['ponde'][i]
    
    coords = []
    for j in range(0, len(all_coords[0])):
        temp_coord = [all_coords[0][j][1], all_coords[0][j][0]]
        coord_tuple = tuple(temp_coord)
        coords.append(coord_tuple)
        
    agebs = zip(*coords)
    gmap.polygon(*agebs, color='#D21404', edge_width=1, face_alpha=ponde)


gmap.draw('mkt_routes_maps.html')
```

# Denue

```python
import requests
import time
```

```python
denue = df_geo_final.copy()
```

```python
token = ''
```

```python
denue.head(1)
```

```python
denue['new_id'] = 1
denue['new_id'] = denue['new_id'].cumsum()
```

```python
denue['new_id'].describe()
```

```python
denue_data = pd.DataFrame()
```

```python
denue_data = pd.read_csv('../data/raw/denue_data.csv')
```

```python
denue.new_id.max()
```

```python
denue_data.new_id.max()
```

```python
%%time

for i in range(2000, 2120):
    temp = denue[denue.new_id==i]
    lat = temp.lat.values[0]
    lon = temp.lon.values[0]
    
    url = f'https://www.inegi.org.mx/app/api/denue/v1/consulta/Buscar/todos/{lat},{lon}/250/{token}'
    
    try:
        response = requests.request("GET", url)
        r = response.json()
        df_temp = pd.DataFrame(r)
        df_temp['new_id'] = i
        denue_data = pd.concat([denue_data,df_temp])
        time.sleep(1)

    except:
        None
    
```

```python
len(denue_data)
```

```python
denue_data.to_csv('../data/raw/denue_data.csv', index=False)
```

```python
len(denue_data)
```

```python
denue_data['Clase_actividad'] = denue_data['Clase_actividad'].str.lower()
```

```python
denue_data['is_consultorio'] = np.where(
    denue_data.Clase_actividad.str.contains('consultorio|laboratorios'), 1, 0
)
denue_data['is_recreativo'] = np.where(
    denue_data.Clase_actividad.str.contains('restaurante|cafeter|bares'), 1, 0
)
denue_data['is_retail'] = np.where(
    denue_data.Clase_actividad.str.contains('belleza|al por menor|lavanderias'), 1, 0
)
denue_data['is_escuela'] = np.where(
    denue_data.Clase_actividad.str.contains('escuelas'), 1, 0
)
```

```python
# Hay que reducir el df para poderlo pegar al original
denue_data_g = denue_data.groupby('new_id').agg({'Id':'count',
                              'is_consultorio':'sum',
                              'is_recreativo':'sum',
                              'is_retail':'sum',
                              'is_escuela':'sum',
                             }).reset_index()
```

```python
denue_data_g = denue_data_g.rename(columns={'Id':'num_comercios'})
```

```python
denue_data_g.new_id.describe()
```

```python
df_denue_geo = denue.merge(denue_data_g, how='left', on='new_id')
```

```python
df_denue_geo.to_csv('../data/processed/denue_geo_data.csv', index=False)
```
